We have a total of 231637 recipes and each recipe is described by 12 variables. 
- id entier Identifiant unique de la recette
- name texte Nom de la recette
- minutes numérique (int) Temps total estimé pour la recette (préparation + cuisson)
- contributor_id entier Identifiant de l’utilisateur ayant publié la recette
- submitted date Date de soumission de la recette
- tags liste (chaîne de type "['tag1','tag2']")
- nutrition liste (chaîne de type "[val1, val2, ...]") Données nutritionnelles (généralement : calories, gras, sucre, sodium, protéines, etc.)
- n_steps entier Nombre d’étapes dans la recette
- steps liste (chaîne list-like) Description détaillée des étapes
- description texte Description libre de la recette
- ingredients liste (chaîne list-like) Liste des ingrédients
- n_ingredients entier Nombre d’ingrédients utilisés


#### 3.2 Qualité des Données

```python
#  ANALYSE DES VALEURS MANQUANTES
print("VALEURS MANQUANTES DANS LES RECETTES")
print("=" * 45)
missing_recipes = recipes.isnull().sum()
total_recipes = len(recipes)

for col in missing_recipes.index:
    missing_count = missing_recipes[col]
    missing_pct = (missing_count / total_recipes) * 100
    if missing_count > 0:
        print(f"   • {col}: {missing_count:,} ({missing_pct:.1f}%)")
    else:
        print(f"    {col}: Aucune valeur manquante")

print(f"\n VALEURS MANQUANTES DANS LES INTERACTIONS")
print("=" * 48)
missing_interactions = interactions.isnull().sum()
total_interactions = len(interactions)

for col in missing_interactions.index:
    missing_count = missing_interactions[col]
    missing_pct = (missing_count / total_interactions) * 100
    if missing_count > 0:
        print(f"   • {col}: {missing_count:,} ({missing_pct:.1f}%)")
    else:
        print(f"    {col}: Aucune valeur manquante")
```

```python
#  DÉTECTION DES DOUBLONS
print(" ANALYSE DES DOUBLONS")
print("=" * 30)

# Doublons dans les recettes
recipes_duplicates = recipes.duplicated().sum()
print(f" Recettes dupliquées: {recipes_duplicates:,}")

if recipes_duplicates > 0:
    # Analyser les doublons par colonne
    recipes_dup_by_col = recipes.duplicated(subset=['name']).sum()
    print(f"   • Doublons par nom: {recipes_dup_by_col:,}")

# Doublons dans les interactions
interactions_duplicates = interactions.duplicated().sum()
print(f"\n Interactions dupliquées: {interactions_duplicates:,}")

if interactions_duplicates > 0:
    # Analyser les doublons par combinaison
    interactions_dup_user_recipe = interactions.duplicated(subset=['user_id', 'recipe_id']).sum()
    print(f"   • Doublons user_id + recipe_id: {interactions_dup_user_recipe:,}")
```
